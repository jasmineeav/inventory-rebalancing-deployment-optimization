```python
# =========================================
# product_master. csv
# =========================================
import pandas as pd
import random

random.seed(42)

brand = "Ocean Spray"

# We want exactly 20 SKUs for 10oz
ten_oz_flavors = [
    "Cranberry",
    "Cranberry Grape",
    "Cranberry Apple",
    "White Cranberry",
    "Diet Cranberry",
    "Cranberry Pineapple",
    "Cranberry Mango",
    "Cranberry Raspberry",
    "Cranberry Cherry",
    "Cranberry Lemonade"
]

ten_oz_types = ["Juice", "Diet Juice"]
pack_types = ["Bottle", "Case"]

# Build 20 exact 10oz SKUs
ten_oz_rows = []
sku_counter = 1

for flavor in ten_oz_flavors:
    for ptype in ten_oz_types:
        sku_id = f"SKU{sku_counter:04d}"
        ten_oz_rows.append({
            "sku_id": sku_id,
            "sku_desc": f"{flavor} {ptype} 10oz Bottle",
            "brand": brand,
            "report_size": "10oz",
            "uom": "CASE",
            "product_type": ptype,
            "flavor": flavor,
            "pack_type": "Bottle",
            "shelf_life_days": 365 if ptype == "Juice" else 300,
            "is_active": "Y"
        })
        sku_counter += 1

ten_oz_df = pd.DataFrame(ten_oz_rows)
print("10oz SKU count:", len(ten_oz_df))
ten_oz_df.head()
```

```python
# =========================================
# generate 200 skus 
# =========================================
other_sizes = ["15.2oz", "64oz", "96oz"]
other_types = ["Juice", "Diet Juice", "Blend", "Cocktail"]
other_flavors = [
    "Cranberry",
    "Cranberry Grape",
    "Cranberry Apple",
    "White Cranberry",
    "Diet Cranberry",
    "Cranberry Pineapple",
    "Cranberry Mango",
    "Cranberry Raspberry",
    "Cranberry Cherry",
    "Cranberry Lemonade",
    "Ruby Cranberry",
    "Cran Tropical",
    "Cran Energy",
    "Cran Citrus",
    "Cranberry Peach"
]
other_pack_types = ["Bottle", "PET", "Case"]

rows = []

while len(rows) < 200:
    size = random.choice(other_sizes)
    ptype = random.choice(other_types)
    flavor = random.choice(other_flavors)
    pack = random.choice(other_pack_types)

    sku_id = f"SKU{sku_counter:04d}"
    sku_desc = f"{flavor} {ptype} {size} {pack}"

    if ptype == "Juice":
        shelf_life_days = 365
    elif ptype == "Diet Juice":
        shelf_life_days = 300
    elif ptype == "Blend":
        shelf_life_days = 270
    else:
        shelf_life_days = 365

    rows.append({
        "sku_id": sku_id,
        "sku_desc": sku_desc,
        "brand": brand,
        "report_size": size,
        "uom": "CASE",
        "product_type": ptype,
        "flavor": flavor,
        "pack_type": pack,
        "shelf_life_days": shelf_life_days,
        "is_active": random.choice(["Y", "Y", "Y", "Y", "N"])
    })
    sku_counter += 1

other_df = pd.DataFrame(rows)

product_master = pd.concat([ten_oz_df, other_df], ignore_index=True)

print("Total SKU count:", len(product_master))
print(product_master["report_size"].value_counts())
product_master.head()
```

```python
# rebuild weekly_fcst.csv
# 定义基础维度
dc_mapping = {
    "PA": "Pennsylvania DC",
    "TX": "Texas DC",
    "FL": "Florida DC",
    "NV": "Nevada DC",
    "WI": "Wisconsin DC",
    "WA": "Washington DC"
}

customers = ["Walmart", "Amazon", "Sam's", "Target", "Costco"]
forecast_weeks = [f"2026-W{str(i).zfill(2)}" for i in range(15, 27)]  # W15-W26
```

```python
# 一个更可控的 forecast 生成规则: 这次我们让不同分组的 Walmart 10oz SKU 需求强度不同：
# healthy: 中等
# incoming_supply_cover: 中高
# rebalancing_needed: 高
# true_supply_risk: 高

# 同时对某些目标 DC 故意放大 demand，方便后面做 deployment。
def get_base_fcst_non_walmart(report_size):
    if report_size == "10oz":
        return random.randint(5000, 12000)
    elif report_size == "15.2oz":
        return random.randint(7000, 18000)
    elif report_size == "64oz":
        return random.randint(10000, 26000)
    else:
        return random.randint(9000, 22000)

def get_base_fcst_for_group(group_name):
    if group_name == "healthy":
        return random.randint(7000, 11000)
    elif group_name == "incoming_supply_cover":
        return random.randint(10000, 14000)
    elif group_name == "rebalancing_needed":
        return random.randint(12000, 18000)
    else:  # true_supply_risk
        return random.randint(13000, 19000)

customer_multiplier = {
    "Walmart": 1.15,
    "Amazon": 0.85,
    "Sam's": 1.00,
    "Target": 0.92,
    "Costco": 1.08
}

dc_multiplier = {
    "PA": 1.08,
    "TX": 1.05,
    "FL": 1.00,
    "NV": 0.95,
    "WI": 0.90,
    "WA": 0.88
}
```

```python
#先定义 Walmart 18 个 SKU 的重点 DC 逻辑
# 这一步是为了后面 deployment：有些 rebalancing_needed SKU，让 PA 高库存、FL 低库存高需求 或者 PA 高库存、NV 极低库存高需求 先定义一个目标 DC map：
rebalancing_target_map = {}
for sku in rebalancing_skus:
    rebalancing_target_map[sku] = random.choice(["FL", "NV"])

true_risk_target_map = {}
for sku in true_risk_skus:
    true_risk_target_map[sku] = random.choice(["FL", "NV", "WA"])

print("rebalancing_target_map:", rebalancing_target_map)
print("true_risk_target_map:", true_risk_target_map)
```

```python
# 生成fcst
rows = []

deployment_group_lookup = {
    **{sku: "healthy" for sku in healthy_skus},
    **{sku: "incoming_supply_cover" for sku in incoming_cover_skus},
    **{sku: "rebalancing_needed" for sku in rebalancing_skus},
    **{sku: "true_supply_risk" for sku in true_risk_skus},
}

for _, sku_row in product_master.iterrows():
    sku_id = sku_row["sku_id"]
    report_size = sku_row["report_size"]
    uom = sku_row["uom"]

    # Case 1: Walmart 10oz scoped SKUs
    if sku_id in walmart_10oz_skus:
        group_name = deployment_group_lookup[sku_id]
        assigned_customers = ["Walmart"]

        # add one optional secondary customer for realism
        if random.random() < 0.35:
            assigned_customers += random.sample(
                [c for c in customers if c != "Walmart"], 1
            )

        base_fcst = get_base_fcst_for_group(group_name)

        # assign DCs
        if group_name == "healthy":
            assigned_dcs = ["PA", "TX", "FL", "NV"]
        elif group_name == "incoming_supply_cover":
            assigned_dcs = ["PA", "FL", "NV", "WA"]
        elif group_name == "rebalancing_needed":
            target_dc = rebalancing_target_map[sku_id]
            donor_dc = "PA"
            assigned_dcs = list(set([donor_dc, target_dc, "TX", "WI"]))
        else:
            target_dc = true_risk_target_map[sku_id]
            assigned_dcs = list(set([target_dc, "FL", "WA", "TX"]))

        for customer in assigned_customers:
            for dc_id in assigned_dcs:
                dc_name = dc_mapping[dc_id]

                # base with customer + dc multiplier
                dc_base = base_fcst * customer_multiplier[customer] * dc_multiplier[dc_id]

                # make target DC demand higher for risk / deployment groups
                if group_name == "rebalancing_needed" and dc_id == rebalancing_target_map[sku_id]:
                    dc_base *= 1.35

                if group_name == "true_supply_risk" and dc_id == true_risk_target_map[sku_id]:
                    dc_base *= 1.40

                # donor PA sometimes a bit lower demand
                if group_name == "rebalancing_needed" and dc_id == "PA":
                    dc_base *= 0.85

                trend_type = random.choice(["flat", "up", "volatile"])

                for week_idx, forecast_week in enumerate(forecast_weeks):
                    if trend_type == "flat":
                        trend_factor = 1.0
                    elif trend_type == "up":
                        trend_factor = 1 + week_idx * 0.012
                    else:
                        trend_factor = random.uniform(0.93, 1.10)

                    noise = random.uniform(0.95, 1.06)
                    forecast_qty = int(dc_base * trend_factor * noise)

                    rows.append({
                        "forecast_week": forecast_week,
                        "sku_id": sku_id,
                        "dc_id": dc_id,
                        "dc_name": dc_name,
                        "customer": customer,
                        "uom": uom,
                        "forecast_qty": max(forecast_qty, 0)
                    })

    # Case 2: all other SKUs
    else:
        assigned_customers = random.sample(customers, k=random.choice([1, 2, 3]))
        base_fcst = get_base_fcst_non_walmart(report_size)

        for customer in assigned_customers:
            assigned_dcs = random.sample(list(dc_mapping.keys()), k=random.choice([2, 3, 4, 5]))

            for dc_id in assigned_dcs:
                dc_name = dc_mapping[dc_id]
                dc_base = base_fcst * customer_multiplier[customer] * dc_multiplier[dc_id]
                trend_type = random.choice(["flat", "up", "down", "volatile"])

                for week_idx, forecast_week in enumerate(forecast_weeks):
                    if trend_type == "flat":
                        trend_factor = 1.0
                    elif trend_type == "up":
                        trend_factor = 1 + week_idx * 0.010
                    elif trend_type == "down":
                        trend_factor = 1 - week_idx * 0.008
                    else:
                        trend_factor = random.uniform(0.92, 1.09)

                    noise = random.uniform(0.94, 1.07)
                    forecast_qty = int(dc_base * trend_factor * noise)

                    rows.append({
                        "forecast_week": forecast_week,
                        "sku_id": sku_id,
                        "dc_id": dc_id,
                        "dc_name": dc_name,
                        "customer": customer,
                        "uom": uom,
                        "forecast_qty": max(forecast_qty, 0)
                    })

weekly_forecast = pd.DataFrame(rows)
print("weekly_forecast shape:", weekly_forecast.shape)
weekly_forecast.head()
```

```python
#rebuild inventory_snapshot.csv
import pandas as pd
import random

random.seed(42)

snapshot_date = "2026-04-10"

dc_mapping = {
    "PA": "Pennsylvania DC",
    "TX": "Texas DC",
    "FL": "Florida DC",
    "NV": "Nevada DC",
    "WI": "Wisconsin DC",
    "WA": "Washington DC"
}

# 先做一个 helper：从 forecast 里算未来4周平均周需求
forecast_4w = weekly_forecast[weekly_forecast["forecast_week"].isin(["2026-W15", "2026-W16", "2026-W17", "2026-W18"])].copy()

avg_fcst_4w = (
    forecast_4w
    .groupby(["sku_id", "dc_id"], as_index=False)
    .agg(avg_weekly_forecast_4w=("forecast_qty", "mean"))
)

# 为了后面查方便，转成 dict
fcst_lookup = {
    (row["sku_id"], row["dc_id"]): row["avg_weekly_forecast_4w"]
    for _, row in avg_fcst_4w.iterrows()
}

rows = []

for _, sku_row in product_master.iterrows():
    sku_id = sku_row["sku_id"]
    report_size = sku_row["report_size"]
    product_type = sku_row["product_type"]

    # 给普通SKU先一个基础库存倍数（相对 forecast）
    for dc_id, dc_name in dc_mapping.items():
        avg_fcst = fcst_lookup.get((sku_id, dc_id), 0)

        # 如果这个 SKU-DC 根本没有 forecast，就给一个小库存或零库存更真实
        if avg_fcst == 0:
            unrestricted_inventory = random.randint(0, 5000)
            blocked_inventory = int(unrestricted_inventory * random.uniform(0.00, 0.05))
            qi_inventory = int(unrestricted_inventory * random.uniform(0.00, 0.08))

            rows.append({
                "snapshot_date": snapshot_date,
                "dc_id": dc_id,
                "dc_name": dc_name,
                "sku_id": sku_id,
                "unrestricted_inventory": unrestricted_inventory,
                "blocked_inventory": blocked_inventory,
                "qi_inventory": qi_inventory
            })
            continue

        # 默认逻辑：普通SKU按 2.5~6 周库存
        target_wos = random.uniform(2.5, 6.0)

        # 对 Walmart 10oz 四类组做特别控制
        if sku_id in healthy_skus:
            # 全网大多健康，PA/TX 更高
            if dc_id in ["PA", "TX"]:
                target_wos = random.uniform(6.0, 10.0)
            else:
                target_wos = random.uniform(3.5, 6.5)

        elif sku_id in incoming_cover_skus:
            # 某些DC偏低，但不是极端低
            if dc_id in ["FL", "NV", "WA"]:
                target_wos = random.uniform(1.2, 2.0)
            else:
                target_wos = random.uniform(3.0, 5.5)

        elif sku_id in rebalancing_skus:
            target_dc = rebalancing_target_map[sku_id]

            # donor DC 固定给 PA
            if dc_id == "PA":
                target_wos = random.uniform(8.0, 13.0)
            elif dc_id == target_dc:
                target_wos = random.uniform(0.4, 1.2)
            elif dc_id in ["TX", "WI"]:
                target_wos = random.uniform(2.5, 4.5)
            else:
                target_wos = random.uniform(1.8, 3.5)

        elif sku_id in true_risk_skus:
            target_dc = true_risk_target_map[sku_id]

            # target DC 很低，全网其他 DC 也不算高
            if dc_id == target_dc:
                target_wos = random.uniform(0.2, 0.8)
            elif dc_id == "PA":
                target_wos = random.uniform(2.0, 3.5)
            else:
                target_wos = random.uniform(1.0, 2.5)

        else:
            # 其他非Walmart重点SKU：正常背景数据
            if dc_id in ["PA", "TX"]:
                target_wos = random.uniform(3.0, 6.0)
            else:
                target_wos = random.uniform(2.0, 5.0)

        unrestricted_inventory = int(avg_fcst * target_wos)

        # 增加一点随机波动
        unrestricted_inventory = int(unrestricted_inventory * random.uniform(0.95, 1.08))

        # blocked / qi 通常较小
        blocked_inventory = int(unrestricted_inventory * random.uniform(0.00, 0.06))
        qi_inventory = int(unrestricted_inventory * random.uniform(0.00, 0.10))

        rows.append({
            "snapshot_date": snapshot_date,
            "dc_id": dc_id,
            "dc_name": dc_name,
            "sku_id": sku_id,
            "unrestricted_inventory": max(unrestricted_inventory, 0),
            "blocked_inventory": max(blocked_inventory, 0),
            "qi_inventory": max(qi_inventory, 0)
        })

inventory_snapshot = pd.DataFrame(rows)

inventory_snapshot_file = raw_path / "inventory_snapshot.csv"
inventory_snapshot.to_csv(inventory_snapshot_file, index=False)

print("Saved:", inventory_snapshot_file)
print("Shape:", inventory_snapshot.shape)
inventory_snapshot.head()
```

```python
# 定义 primary plant
import pandas as pd
import random

random.seed(42)

plant_state_map = {
    "LHV": "PA",
    "Henderson": "NV",
    "Sulphur": "TX"
}

plant_lines = {
    "LHV": ["L1", "L2", "L3"],
    "Henderson": ["H1", "H2", "H3"],
    "Sulphur": ["S1", "S2", "S3"]
}

production_weeks = [f"2026-W{str(i).zfill(2)}" for i in range(15, 27)]  # W15-W26
frozen_weeks = ["2026-W15", "2026-W16", "2026-W17", "2026-W18"]
```

```python
# 给 SKU 分 primary plant
primary_plant_map = {}

for sku in product_master["sku_id"].tolist():
    # Rebalancing group
    if sku in rebalancing_skus:
        target_dc = rebalancing_target_map[sku]
        if target_dc == "NV":
            primary_plant_map[sku] = "Henderson"
        elif target_dc == "FL":
            primary_plant_map[sku] = random.choice(["LHV", "Sulphur"])
        else:
            primary_plant_map[sku] = random.choice(["LHV", "Sulphur", "Henderson"])

    # True risk group
    elif sku in true_risk_skus:
        target_dc = true_risk_target_map[sku]
        if target_dc == "NV":
            primary_plant_map[sku] = "Henderson"
        elif target_dc == "TX":
            primary_plant_map[sku] = "Sulphur"
        else:
            primary_plant_map[sku] = random.choice(["LHV", "Sulphur", "Henderson"])

    # Incoming cover group
    elif sku in incoming_cover_skus:
        primary_plant_map[sku] = random.choice(["LHV", "Sulphur", "Henderson"])

    # Healthy group
    elif sku in healthy_skus:
        primary_plant_map[sku] = random.choice(["LHV", "Sulphur", "Henderson"])

    # Other background SKUs
    else:
        primary_plant_map[sku] = random.choice(["LHV", "Sulphur", "Henderson"])

primary_plant_df = pd.DataFrame({
    "sku_id": list(primary_plant_map.keys()),
    "primary_plant": list(primary_plant_map.values())
})

primary_plant_df["plant_state"] = primary_plant_df["primary_plant"].map(plant_state_map)
primary_plant_df.head()
```

```python
#定义 base production qty
def get_base_production_qty(report_size, product_type):
    if report_size == "10oz":
        base = random.randint(30000, 90000)
    elif report_size == "15.2oz":
        base = random.randint(50000, 130000)
    elif report_size == "64oz":
        base = random.randint(70000, 180000)
    else:  # 96oz
        base = random.randint(60000, 160000)

    if product_type == "Juice":
        base *= 1.05
    elif product_type == "Diet Juice":
        base *= 0.90
    elif product_type == "Blend":
        base *= 1.00
    else:  # Cocktail
        base *= 0.95

    return int(base)
```

```python
# 生成 production plan（核心）这一步会把 cadence 和 constraint 都写进去。
rows = []

product_lookup = product_master.set_index("sku_id").to_dict("index")

for sku in product_master["sku_id"].tolist():
    report_size = product_lookup[sku]["report_size"]
    product_type = product_lookup[sku]["product_type"]
    uom = product_lookup[sku]["uom"]

    plant_id = primary_plant_map[sku]
    plant_state = plant_state_map[plant_id]
    line = random.choice(plant_lines[plant_id])

    cadence = cadence_map[sku]
    base_qty = get_base_production_qty(report_size, product_type)

    # Decide which weeks this SKU produces
    if cadence == "weekly":
        active_weeks = production_weeks.copy()

    elif cadence == "biweekly":
        # choose one alternating pattern
        if random.random() < 0.5:
            active_weeks = ["2026-W15", "2026-W17", "2026-W19", "2026-W21", "2026-W23", "2026-W25"]
        else:
            active_weeks = ["2026-W16", "2026-W18", "2026-W20", "2026-W22", "2026-W24", "2026-W26"]

    else:  # monthly
        # choose one every-4-week pattern
        monthly_patterns = [
            ["2026-W15", "2026-W19", "2026-W23"],
            ["2026-W16", "2026-W20", "2026-W24"],
            ["2026-W17", "2026-W21", "2026-W25"],
            ["2026-W18", "2026-W22", "2026-W26"]
        ]
        active_weeks = random.choice(monthly_patterns)

    # Special constraints by deployment group
    # Healthy → normal supply
    # Incoming cover → at least one frozen-week production
    # Rebalancing → frozen production exists but often weaker or later
    # True risk → weak or no frozen production

    if sku in incoming_cover_skus:
        # ensure at least one frozen week is present
        if not any(w in frozen_weeks for w in active_weeks):
            active_weeks = active_weeks[:-1] + [random.choice(frozen_weeks)]

    if sku in rebalancing_skus:
        # make frozen support weaker overall
        pass

    if sku in true_risk_skus:
        # force frozen horizon to be weak or absent
        active_weeks = [w for w in active_weeks if w not in frozen_weeks]

        # if all removed, keep only week5+ production
        if len(active_weeks) == 0:
            active_weeks = ["2026-W19", "2026-W23"]

    for week in production_weeks:
        if week not in active_weeks:
            continue

        qty = base_qty

        # frozen vs non-frozen variability
        if week in frozen_weeks:
            variability = random.uniform(0.96, 1.04)
        else:
            variability = random.uniform(0.88, 1.12)

        qty = int(qty * variability)

        # Group-specific production strength
        if sku in healthy_skus:
            qty = int(qty * random.uniform(0.95, 1.08))

        elif sku in incoming_cover_skus:
            # enough to partially or mostly cover
            qty = int(qty * random.uniform(0.75, 0.95))

        elif sku in rebalancing_skus:
            # not zero, but not strong enough to rescue target DC in time
            if week in frozen_weeks:
                qty = int(qty * random.uniform(0.30, 0.55))
            else:
                qty = int(qty * random.uniform(0.65, 0.90))

        elif sku in true_risk_skus:
            # weak, and mostly outside frozen horizon
            if week in frozen_weeks:
                qty = int(qty * random.uniform(0.05, 0.20))
            else:
                qty = int(qty * random.uniform(0.35, 0.60))

        else:
            qty = int(qty * random.uniform(0.80, 1.05))

        # simulate line occupied / no extra capacity for certain true-risk NV cases
        if sku in true_risk_skus and plant_id == "Henderson" and week in frozen_weeks:
            qty = int(qty * random.uniform(0.00, 0.10))

        rows.append({
            "production_week": week,
            "plant_id": plant_id,
            "plant_state": plant_state,
            "sku_id": sku,
            "production_line": line,
            "planned_qty": max(qty, 0),
            "uom": uom
        })

production_plan = pd.DataFrame(rows)

production_plan_file = raw_path / "production_plan.csv"
production_plan.to_csv(production_plan_file, index=False)

print("Saved:", production_plan_file)
print("Shape:", production_plan.shape)
production_plan.head()
```

```python
## ----------SQL Risk Model + Deployment Recommendation------------------
#定义scope
con.execute("""
CREATE OR REPLACE VIEW sku_scope AS
SELECT DISTINCT
    wf.sku_id,
    wf.customer,
    pm.report_size,
    pm.sku_desc,
    pm.uom
FROM weekly_forecast wf
JOIN product_master pm
    ON wf.sku_id = pm.sku_id
WHERE wf.customer = 'Walmart'
  AND pm.report_size = '10oz';
""")
```

```python
#算未来 4 周 demand
con.execute("""
CREATE OR REPLACE VIEW forecast_4w AS
SELECT
    wf.sku_id,
    wf.dc_id,
    wf.dc_name,
    wf.customer,
    wf.forecast_week,
    wf.forecast_qty
FROM weekly_forecast wf
JOIN sku_scope s
    ON wf.sku_id = s.sku_id
   AND wf.customer = s.customer
WHERE wf.forecast_week IN ('2026-W15','2026-W16','2026-W17','2026-W18');
""")
```

```python
con.execute("""
CREATE OR REPLACE VIEW forecast_4w_summary AS
SELECT
    sku_id,
    dc_id,
    dc_name,
    customer,
    SUM(forecast_qty) AS forecast_4w_qty,
    AVG(forecast_qty) AS avg_weekly_forecast
FROM forecast_4w
GROUP BY sku_id, dc_id, dc_name, customer;
""")
```

```python
con.execute("""
CREATE OR REPLACE VIEW inventory_wos AS
SELECT
    f.sku_id,
    f.dc_id,
    f.dc_name,
    f.customer,
    i.snapshot_date,
    i.unrestricted_inventory,
    i.blocked_inventory,
    i.qi_inventory,
    f.forecast_4w_qty,
    f.avg_weekly_forecast,
    CASE
        WHEN f.avg_weekly_forecast > 0
        THEN ROUND(i.unrestricted_inventory * 1.0 / f.avg_weekly_forecast, 2)
        ELSE NULL
    END AS wos
FROM forecast_4w_summary f
LEFT JOIN inventory_snapshot i
    ON f.sku_id = i.sku_id
   AND f.dc_id = i.dc_id;
""")
```

```python
#先看 low WOS
con.sql("""
SELECT *
FROM inventory_wos
WHERE wos < 2
ORDER BY wos ASC
LIMIT 30
""").df()
```

```python
#算 frozen production support
con.execute("""
CREATE OR REPLACE VIEW production_frozen_4w AS
SELECT
    sku_id,
    plant_id,
    plant_state,
    production_week,
    planned_qty
FROM production_plan
WHERE production_week IN ('2026-W15','2026-W16','2026-W17','2026-W18');
""")
```

```python
# 汇总SKU
con.execute("""
CREATE OR REPLACE VIEW production_frozen_summary AS
SELECT
    sku_id,
    SUM(planned_qty) AS frozen_4w_production_qty
FROM production_frozen_4w
GROUP BY sku_id;
""")
```

```python
# 再做一个“保守的 DC support 估算”。 因为 plant 产量不可能完整给每个 DC，我们先取 15% 作为潜在可分配量：
con.execute("""
CREATE OR REPLACE VIEW production_dc_support_est AS
SELECT
    sku_id,
    ROUND(frozen_4w_production_qty * 0.15, 0) AS est_dc_production_support
FROM production_frozen_summary;
""")
```

```python
#算 STO inbound
con.execute("""
CREATE OR REPLACE VIEW sto_inbound_4w AS
SELECT
    sku_id,
    to_node AS dc_id,
    SUM(shipment_qty) AS inbound_sto_qty
FROM sto
WHERE pgr_date IS NULL
  AND should_deliver_date <= DATE '2026-05-08'
GROUP BY sku_id, to_node;
""")
```

```python
# 找 donor DC 这个是 deployment 项目的关键。
# 我们定义 donor 条件： wos > 6 且 unrestricted_inventory - 2 * avg_weekly_forecast > 0
# 也就是这个 DC 自己很健康，并且有可转出的余量。
con.execute("""
CREATE OR REPLACE VIEW donor_candidates AS
SELECT
    sku_id,
    dc_id AS donor_dc,
    dc_name AS donor_dc_name,
    unrestricted_inventory,
    avg_weekly_forecast,
    wos,
    ROUND(unrestricted_inventory - 2 * avg_weekly_forecast, 0) AS donor_surplus_qty
FROM inventory_wos
WHERE wos > 6
  AND (unrestricted_inventory - 2 * avg_weekly_forecast) > 0;
""")
```

```python
# 看一下donor
con.sql("""
SELECT *
FROM donor_candidates
ORDER BY sku_id, wos DESC
LIMIT 30
""").df()
```

```python
#先做最终风险分类
# 逻辑如下： Covered Risk wos < 2 且有 STO inbound
# 或 有 production support Rebalancing Opportunity wos < 2 没有 STO production support 也不够 但存在 donor DC True Supply Risk wos < 2 没有 STO production support 不够 也没有 donor DC

# 先建一个 donor flag：
con.execute("""
CREATE OR REPLACE VIEW donor_flag AS
SELECT
    sku_id,
    COUNT(*) AS donor_dc_count
FROM donor_candidates
GROUP BY sku_id;
""")
```

```python
#最终风险表：
con.execute("""
CREATE OR REPLACE VIEW risk_final AS
SELECT
    i.sku_id,
    i.dc_id,
    i.dc_name,
    i.customer,
    i.unrestricted_inventory,
    i.avg_weekly_forecast,
    i.wos,
    COALESCE(s.inbound_sto_qty, 0) AS inbound_sto_qty,
    COALESCE(p.est_dc_production_support, 0) AS est_dc_production_support,
    COALESCE(d.donor_dc_count, 0) AS donor_dc_count,

    CASE
        WHEN i.wos >= 2 THEN 'Healthy'

        WHEN i.wos < 2
             AND COALESCE(s.inbound_sto_qty, 0) >= i.avg_weekly_forecast
        THEN 'Covered Risk'

        WHEN i.wos < 2
             AND COALESCE(p.est_dc_production_support, 0) >= i.avg_weekly_forecast
        THEN 'Covered Risk'

        WHEN i.wos < 2
             AND COALESCE(d.donor_dc_count, 0) > 0
        THEN 'Rebalancing Opportunity'

        ELSE 'True Supply Risk'
    END AS final_risk
FROM inventory_wos i
LEFT JOIN sto_inbound_4w s
    ON i.sku_id = s.sku_id
   AND i.dc_id = s.dc_id
LEFT JOIN production_dc_support_est p
    ON i.sku_id = p.sku_id
LEFT JOIN donor_flag d
    ON i.sku_id = d.sku_id;
""")
```

```python
# 看分类结果
con.sql("""
SELECT
    final_risk,
    COUNT(*) AS cnt
FROM risk_final
GROUP BY final_risk
ORDER BY cnt DESC
""").df()
```

```python
# 做 deployment recommendation
# 我们先简单做法： 对每个 sku_id，选 wos 最高的 donor DC。
con.execute("""
CREATE OR REPLACE VIEW best_donor AS
SELECT *
FROM (
    SELECT
        sku_id,
        donor_dc,
        donor_dc_name,
        donor_surplus_qty,
        wos,
        ROW_NUMBER() OVER (
            PARTITION BY sku_id
            ORDER BY wos DESC, donor_surplus_qty DESC
        ) AS rn
    FROM donor_candidates
) t
WHERE rn = 1;
""")
```

```python
# 输出 recommendation：
con.execute("""
CREATE OR REPLACE VIEW deployment_recommendation AS
SELECT
    r.sku_id,
    r.dc_id AS shortage_dc,
    r.dc_name AS shortage_dc_name,
    r.wos AS shortage_wos,
    r.unrestricted_inventory,
    r.avg_weekly_forecast,
    ROUND(2 * r.avg_weekly_forecast - r.unrestricted_inventory, 0) AS recommended_transfer_qty,
    b.donor_dc,
    b.donor_dc_name,
    b.donor_surplus_qty,
    r.final_risk,
    CASE
        WHEN r.final_risk = 'Rebalancing Opportunity'
        THEN 'Transfer inventory from donor DC to shortage DC'
        WHEN r.final_risk = 'Covered Risk'
        THEN 'Monitor inbound supply / production'
        WHEN r.final_risk = 'True Supply Risk'
        THEN 'Escalate to planner: no feasible donor within network'
        ELSE 'No action required'
    END AS recommendation
FROM risk_final r
LEFT JOIN best_donor b
    ON r.sku_id = b.sku_id
WHERE r.final_risk IN ('Rebalancing Opportunity', 'Covered Risk', 'True Supply Risk');
""")
```

```python
# Phase 5 — Time-Phased Inventory Projection & Replenishment Feasibility
# 这一步要做什么

# 对每个 sku_id + dc_id，按周滚动：
# Beginning Inventory
# + STO arriving this week
# + Production support arriving this week
# - Forecast this week
# = Projected Ending Inventory

# 然后判断：第几周开始出现 gap， gap 持续几周， STO 能不能在 gap 前补上， Production 能不能在 gap 前补上 是否需要 inter-DC transfer，是否完全无解

# 第 1 步：先把 SQL 结果拉进 pandas
# 先在 notebook 里运行下面这几段，把需要的表读出来。
# forecast by week
forecast_df = con.sql("""
SELECT
    sku_id,
    dc_id,
    dc_name,
    customer,
    forecast_week,
    forecast_qty
FROM forecast_4w
""").df()

forecast_df.head()
```

```python
# 第 2 步：准备辅助映射 我们需要把日期映射到 planning week。
weeks = ["2026-W15", "2026-W16", "2026-W17", "2026-W18"]

week_date_map = {
    "2026-W15": ("2026-04-07", "2026-04-13"),
    "2026-W16": ("2026-04-14", "2026-04-20"),
    "2026-W17": ("2026-04-21", "2026-04-27"),
    "2026-W18": ("2026-04-28", "2026-05-04"),
}

def date_to_week(date_val):
    if pd.isna(date_val):
        return None
    dt = pd.to_datetime(date_val)
    for wk, (start, end) in week_date_map.items():
        if pd.to_datetime(start) <= dt <= pd.to_datetime(end):
            return wk
    return None
```

```python
# 定义 projection scope
# 我们需要的不只是 low-WOS row，还要它们对应 SKU 的 donor rows。
projection_skus = sorted(risk_scope_df["sku_id"].unique().tolist())

projection_inventory_df = inventory_df[inventory_df["sku_id"].isin(projection_skus)].copy()
projection_forecast_df = forecast_df[forecast_df["sku_id"].isin(projection_skus)].copy()
projection_donor_df = donor_df[donor_df["sku_id"].isin(projection_skus)].copy()
projection_sto_df = sto_df[sto_df["sku_id"].isin(projection_skus)].copy()
projection_prod_df = prod_df[prod_df["sku_id"].isin(projection_skus)].copy()

print("projection_skus:", len(projection_skus))
```

```python
# 第 3 步：定义时间映射
# 我们现在只看：
# 2026-W15 2026-W16 2026-W17 2026-W18
# 先定义一个简单的 date → week bucket 函数
import pandas as pd
import numpy as np

projection_weeks = ["2026-W15", "2026-W16", "2026-W17", "2026-W18"]

week_date_map = {
    "2026-W15": ("2026-04-06", "2026-04-12"),
    "2026-W16": ("2026-04-13", "2026-04-19"),
    "2026-W17": ("2026-04-20", "2026-04-26"),
    "2026-W18": ("2026-04-27", "2026-05-03"),
}

def map_date_to_week(date_val):
    if pd.isna(date_val):
        return None
    d = pd.to_datetime(date_val)
    for wk, (start_date, end_date) in week_date_map.items():
        if pd.Timestamp(start_date) <= d <= pd.Timestamp(end_date):
            return wk
    return None
```

```python
# 准备 weekly STO arrivals
projection_sto_df = projection_sto_df.copy()
projection_sto_df["arrival_week"] = projection_sto_df["should_deliver_date"].apply(map_date_to_week)

sto_weekly = (
    projection_sto_df[
        projection_sto_df["arrival_week"].isin(projection_weeks)
    ]
    .groupby(["sku_id", "to_node", "arrival_week"], as_index=False)
    .agg(inbound_sto_qty=("shipment_qty", "sum"))
    .rename(columns={"to_node": "dc_id"})
)

sto_weekly.head()
```

```python
# 准备 weekly production support
# 这里我们先用一个简化但合理的规则：
# 简化规则： production_week 本身就是可用周， 仅把生产量的 15% 视为某 DC 可分配支持， 本州 DC 优先能拿到， 也就是说，先只把 production support 给 plant_state 对应的 DC

# LHV → PA
# Henderson → NV
# Sulphur → TX
projection_prod_df = projection_prod_df.copy()
projection_prod_df["dc_id"] = projection_prod_df["plant_state"]
projection_prod_df["inbound_production_qty"] = (projection_prod_df["planned_qty"] * 0.15).round(0)

prod_weekly = (
    projection_prod_df.groupby(["sku_id", "dc_id", "production_week"], as_index=False)
    .agg(inbound_production_qty=("inbound_production_qty", "sum"))
    .rename(columns={"production_week": "week"})
)

prod_weekly.head()
```

```python
# 准备 beginning inventory base
begin_inv_df = projection_inventory_df.copy()
begin_inv_df = begin_inv_df.rename(columns={"unrestricted_inventory": "begin_inventory"})
begin_inv_df.head()
```

```python
# 建立 rolling projection 主循环

weekly_projection_rows = []

# 建索引方便取数
forecast_lookup = {
    (row["sku_id"], row["dc_id"], row["forecast_week"]): row["forecast_qty"]
    for _, row in projection_forecast_df.iterrows()
}

sto_lookup = {
    (row["sku_id"], row["dc_id"], row["arrival_week"]): row["inbound_sto_qty"]
    for _, row in sto_weekly.iterrows()
}

prod_lookup = {
    (row["sku_id"], row["dc_id"], row["week"]): row["inbound_production_qty"]
    for _, row in prod_weekly.iterrows()
}

for _, base_row in begin_inv_df.iterrows():
    sku_id = base_row["sku_id"]
    dc_id = base_row["dc_id"]
    dc_name = base_row["dc_name"]
    beginning_inventory = base_row["begin_inventory"]

    current_inventory = beginning_inventory

    for week in projection_weeks:
        forecast_qty = forecast_lookup.get((sku_id, dc_id, week), 0)
        inbound_sto_qty = sto_lookup.get((sku_id, dc_id, week), 0)
        inbound_production_qty = prod_lookup.get((sku_id, dc_id, week), 0)

        projected_ending_inventory = (
            current_inventory
            + inbound_sto_qty
            + inbound_production_qty
            - forecast_qty
        )

        projected_wos = (
            projected_ending_inventory / forecast_qty
            if forecast_qty > 0 else None
        )

        gap_flag = 1 if projected_ending_inventory < 0 else 0

        weekly_projection_rows.append({
            "sku_id": sku_id,
            "dc_id": dc_id,
            "dc_name": dc_name,
            "week": week,
            "beginning_inventory": current_inventory,
            "forecast_qty": forecast_qty,
            "inbound_sto_qty": inbound_sto_qty,
            "inbound_production_qty": inbound_production_qty,
            "projected_ending_inventory": projected_ending_inventory,
            "projected_wos": round(projected_wos, 2) if projected_wos is not None else None,
            "gap_flag": gap_flag
        })

        current_inventory = projected_ending_inventory

weekly_inventory_projection = pd.DataFrame(weekly_projection_rows)
weekly_inventory_projection.head(20)
```

```python
# 做 replenishment feasibility

# 这里先做第一版可行性判断：对于每个 gap case：是否已有 STO 在 first gap week 前到达； 是否已有 production support 在 first gap week 前到达； 是否存在 donor DC； 最佳 donor 是谁

#  donor best source：
best_donor_df = (
    projection_donor_df.sort_values(["sku_id", "wos", "donor_surplus_qty"], ascending=[True, False, False])
    .drop_duplicates(subset=["sku_id"])
    .copy()
)

best_donor_df.head()
```

```python
# =========================
# Final replenishment_feasibility
# =========================

# 1) 先选每个 SKU 的最佳 donor
best_donor_df = (
    projection_donor_df
    .sort_values(
        ["sku_id", "wos", "donor_surplus_qty"],
        ascending=[True, False, False]
    )
    .drop_duplicates(subset=["sku_id"])
    .copy()
)

week_order = {wk: i for i, wk in enumerate(projection_weeks)}

replenishment_rows = []

for _, gap_row in gap_summary.iterrows():
    sku_id = gap_row["sku_id"]
    shortage_dc = gap_row["dc_id"]
    shortage_dc_name = gap_row["dc_name"]
    first_gap_week = gap_row["first_gap_week"]
    gap_idx = week_order[first_gap_week]

    max_gap_qty_abs = abs(gap_row["max_gap_qty"])
    cumulative_gap_qty_abs = abs(gap_row["cumulative_gap_qty"])
    total_gap_weeks = gap_row["total_gap_weeks"]

    # ---- support arriving on or before first gap week ----
    sto_qty = sto_weekly[
        (sto_weekly["sku_id"] == sku_id) &
        (sto_weekly["dc_id"] == shortage_dc) &
        (sto_weekly["arrival_week"].map(week_order) <= gap_idx)
    ]["inbound_sto_qty"].sum()

    prod_qty = prod_weekly[
        (prod_weekly["sku_id"] == sku_id) &
        (prod_weekly["dc_id"] == shortage_dc) &
        (prod_weekly["week"].map(week_order) <= gap_idx)
    ]["inbound_production_qty"].sum()

    donor_match = best_donor_df[best_donor_df["sku_id"] == sku_id]

    donor_exists = 1 if len(donor_match) > 0 else 0

    if donor_exists:
        best_donor_dc = donor_match.iloc[0]["donor_dc"]
        best_donor_dc_name = donor_match.iloc[0]["donor_dc_name"]
        donor_surplus_qty = float(donor_match.iloc[0]["donor_surplus_qty"])
        donor_wos = float(donor_match.iloc[0]["wos"])
    else:
        best_donor_dc = None
        best_donor_dc_name = None
        donor_surplus_qty = 0.0
        donor_wos = None

    # ---- quantity targets ----
    # minimum qty to stop the first severe stockout
    qty_to_stabilize = max_gap_qty_abs

    # qty to fully cover cumulative gap across all gap weeks
    qty_to_full_cover = cumulative_gap_qty_abs

    # average weekly gap can be useful as a softer threshold
    avg_gap_per_gap_week = cumulative_gap_qty_abs / total_gap_weeks if total_gap_weeks > 0 else 0

    # ---- stronger coverage logic ----
    # strict cover = can stabilize first gap
    sto_covers_first_gap = sto_qty >= qty_to_stabilize
    prod_covers_first_gap = prod_qty >= qty_to_stabilize

    # softer cover = at least meaningfully offsets the shortage
    sto_partially_covers = sto_qty >= 0.5 * qty_to_stabilize
    prod_partially_covers = prod_qty >= 0.5 * qty_to_stabilize

    donor_can_stabilize = donor_exists and donor_surplus_qty >= qty_to_stabilize
    donor_can_partially_cover = donor_exists and donor_surplus_qty >= avg_gap_per_gap_week

    # ---- decision logic ----
    if sto_covers_first_gap:
        feasible_action_type = "Covered by STO"
        recommended_source = "In-transit STO"
        recommended_qty = None
        business_reason = "Inbound STO arriving by or before the first gap week is sufficient to cover the initial shortage."

    elif prod_covers_first_gap:
        feasible_action_type = "Covered by Production"
        recommended_source = "Primary plant production"
        recommended_qty = None
        business_reason = "Frozen-horizon production support is sufficient to stabilize the first shortage week."

    elif donor_can_stabilize:
        feasible_action_type = "Transfer Required"
        recommended_source = best_donor_dc
        recommended_qty = min(qty_to_full_cover, donor_surplus_qty)
        business_reason = "No timely STO or sufficient production support; donor DC has enough surplus inventory to cover the shortage."

    elif sto_partially_covers or prod_partially_covers or donor_can_partially_cover:
        feasible_action_type = "Partial Transfer / Escalate"

        if donor_exists:
            recommended_source = best_donor_dc
            recommended_qty = min(qty_to_stabilize, donor_surplus_qty)
            business_reason = "Available support can partially mitigate the shortage, but not fully cover the projected gap."
        else:
            recommended_source = None
            recommended_qty = None
            business_reason = "Some supply support exists, but it is insufficient to stabilize the shortage without planner intervention."

    else:
        feasible_action_type = "No Feasible Source"
        recommended_source = None
        recommended_qty = None
        business_reason = "No timely STO, insufficient production support, and no donor DC with enough surplus inventory."

    replenishment_rows.append({
        "sku_id": sku_id,
        "shortage_dc": shortage_dc,
        "shortage_dc_name": shortage_dc_name,
        "first_gap_week": first_gap_week,
        "total_gap_weeks": total_gap_weeks,
        "max_gap_qty_abs": max_gap_qty_abs,
        "cumulative_gap_qty_abs": cumulative_gap_qty_abs,
        "qty_to_stabilize": qty_to_stabilize,
        "qty_to_full_cover": qty_to_full_cover,
        "avg_gap_per_gap_week": avg_gap_per_gap_week,
        "sto_can_cover_qty": sto_qty,
        "production_can_cover_qty": prod_qty,
        "donor_exists": donor_exists,
        "best_donor_dc": best_donor_dc,
        "best_donor_dc_name": best_donor_dc_name,
        "donor_surplus_qty": donor_surplus_qty,
        "donor_wos": donor_wos,
        "recommended_source": recommended_source,
        "feasible_action_type": feasible_action_type,
        "business_reason": business_reason
    })

replenishment_feasibility = pd.DataFrame(replenishment_rows)
replenishment_feasibility.head(20)
```

```python
# 注意： max_gap_qty 这里会是最负的库存值 （ 应该在gap summary 后面） 
# cumulative_gap_qty 也是负值累加
# 如果你想更直观，可以转成正的 gap magnitude：
if len(gap_summary) > 0:
    gap_summary["max_gap_qty_abs"] = gap_summary["max_gap_qty"].abs()
    gap_summary["cumulative_gap_qty_abs"] = gap_summary["cumulative_gap_qty"].abs()
```

```python
# =========================
# Final recommendation
# =========================

final_recommendation = replenishment_feasibility.copy()

# 更清晰的 action text
action_map = {
    "Covered by STO": "Monitor inbound STO and hold deployment",
    "Covered by Production": "Monitor plant replenishment, no transfer needed",
    "Transfer Required": "Transfer inventory from donor DC to shortage DC",
    "Partial Transfer / Escalate": "Execute partial transfer and escalate remaining gap to planner",
    "No Feasible Source": "Escalate to planner: no timely supply or donor available"
}

final_recommendation["recommended_action"] = final_recommendation["feasible_action_type"].map(action_map)

# 两个 quantity 版本：先救火 vs 尽量补平
def calc_recommended_qty_stabilize(row):
    if row["feasible_action_type"] in ["Transfer Required", "Partial Transfer / Escalate"] and row["donor_exists"] == 1:
        return min(row["qty_to_stabilize"], row["donor_surplus_qty"])
    return None

def calc_recommended_qty_full_cover(row):
    if row["feasible_action_type"] == "Transfer Required" and row["donor_exists"] == 1:
        return min(row["qty_to_full_cover"], row["donor_surplus_qty"])
    elif row["feasible_action_type"] == "Partial Transfer / Escalate" and row["donor_exists"] == 1:
        return min(row["qty_to_stabilize"], row["donor_surplus_qty"])
    return None

final_recommendation["recommended_qty_stabilize"] = final_recommendation.apply(
    calc_recommended_qty_stabilize, axis=1
)

final_recommendation["recommended_qty_full_cover"] = final_recommendation.apply(
    calc_recommended_qty_full_cover, axis=1
)

# priority for sorting in BI / reporting
priority_map = {
    "No Feasible Source": 1,
    "Partial Transfer / Escalate": 2,
    "Transfer Required": 3,
    "Covered by Production": 4,
    "Covered by STO": 5
}
final_recommendation["action_priority"] = final_recommendation["feasible_action_type"].map(priority_map)

# 最终输出列
final_recommendation = final_recommendation[[
    "sku_id",
    "shortage_dc",
    "shortage_dc_name",
    "first_gap_week",
    "total_gap_weeks",
    "max_gap_qty_abs",
    "cumulative_gap_qty_abs",
    "qty_to_stabilize",
    "qty_to_full_cover",
    "avg_gap_per_gap_week",
    "sto_can_cover_qty",
    "production_can_cover_qty",
    "donor_exists",
    "best_donor_dc",
    "best_donor_dc_name",
    "donor_surplus_qty",
    "donor_wos",
    "feasible_action_type",
    "recommended_source",
    "recommended_action",
    "recommended_qty_stabilize",
    "recommended_qty_full_cover",
    "business_reason",
    "action_priority"
]].rename(columns={
    "shortage_dc": "dc_id",
    "shortage_dc_name": "dc_name"
})

final_recommendation = final_recommendation.sort_values(
    ["action_priority", "first_gap_week", "cumulative_gap_qty_abs"],
    ascending=[True, True, False]
).reset_index(drop=True)

final_recommendation.head(30)
```
